```python
import numpy as np 
import matplotlib.pyplot as plt

a = np.array([
    [1, 2], 
    [0, 1]
])

b = np.array([
    [2, 0], 
    [3, 4]
])

prod = a @ b

print(prod)

# PLotting product of two arrays

plt.plot(a) 
plt.plot(b) 
plt.plot(prod) 


# Add Title

plt.title("Matplotlib PLotting Product of NumPy Arrays") 

plt.show()
```
